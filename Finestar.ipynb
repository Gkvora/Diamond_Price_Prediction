```python
# the diamond's price depends upon hundreds of parameters but mainly on 4 C's (Carat, Cut, Clarity, Color).
df_=df.drop(['Id','Shape','Polish','Symmetry','Fluorescence','Messurements','Data Url'],axis=1)
```

## Machine Learning Model Building

### Label Encoder converting categorical data into numeric form 

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l1=LabelEncoder()
df_['Cut']=l1.fit_transform(df_['Cut'])
df_['Colour']=l1.fit_transform(df_['Colour'])
df_['Clarity']=l1.fit_transform(df_['Clarity'])
```

```python
df_
```

```python
x=df_.drop(['Price'],axis=1)
y=df_['Price']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
mar = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test,y_pred)
print("mean_absolute_error : ",mar)
print("mean_squared_error : ",mse)
print("root_mean_absolute_error : ",rmse)
print("r2_score : ",r2)
```
